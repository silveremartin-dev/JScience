# üß¨ JScience Grid from Jupyter

This notebook demonstrates how to submit scientific tasks to the JScience Distributed Grid directly from Python.

```python
from jscience import JScienceClient

# Connect to Grid
client = JScienceClient('localhost', 50051)
```

```python
# Authenticate
client.login('admin', 'admin')
```

```python
# Check Grid Status
client.get_status()
```

## Submit Biology Task (DNA Folding)

```python
# Submit a task
task_id = client.submit_task(
    task_type='dna_folding', 
    params={'sequence': 'ATCGGCTA', 'iterations': 1000}, 
    priority='HIGH'
)
```

```python
# Get Result
result = client.wait_for_result(task_id)
result
```
